```python
# Now let's apply liner regression

from sklearn.model_selection import train_test_split

y = df['Ag']
x = df.drop(['Ag'], axis=1)

x.head(),y.head()
```

```python
ones = np.ones((len(x), 1))
x = pd.DataFrame(np.concatenate([x, ones], axis=1))
```

## Now Let's Split the data into a test and training set

```python
X_train, X_test, y_train, y_test = train_test_split(x, y,test_size = .2, random_state=24)
```

```python
np.mean(X_train)
```

```python
# %% Linear Regression with scikit-learn
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(X_train, y_train)

w = np.concatenate([model.coef_, model.intercept_[None]], axis=0)
print(f'w={w}')
```
